# B. SALES DATA
##  Task 1: Merging 12 Months of sales data into a single file

```python
import pandas as pd
import os
```

```python

files = [file for file in os.listdir("sales/sales_data")]

# First Way
dfs = []
for file in files:
    df = pd.read_csv("sales/sales_data/" + file)
    dfs.append(df)
    
df = pd.concat(dfs)
print(df.describe())
print(df.head())


# Second Way
print("________________________________________________________________________________________")

all_months_data = pd.DataFrame()

for file in files:
    df = pd.read_csv("sales/sales_data/" + file)
    all_months_data = pd.concat([all_months_data, df])
    
print(all_months_data.describe())
print(all_months_data.head())



# create combined csv file
all_months_data.to_csv("sales/sales_data/all_data.csv", index=False)
```

## read in updated dataframe

```python
all_data = pd.read_csv("sales/sales_data/all_data.csv")
all_data.head()
```

## Task 2: Add Month Column

```python
# data da bazı sorunlar var duzenlenmesı gerekiyor
all_data["Month"] = all_data["Order Date"].str[0:2].astype("int")
all_data.head()
```

### Clean up the Data
#### 1. drop rows of NAN

```python
nan_df = all_data[all_data.isna().any(axis=1)]
nan_df.head()
```

```python
# eger her hangi bir tanesi nan ise kaldır
# all_data.dropna(how="any")

#eger hepsi NAN ise kaldır
all_data = all_data.dropna(how="all")
all_data.head()
```

#### 2. Find the 'Or' and delete the rows

```python
temp_df = all_data[all_data["Order Date"].str[0:2] == "Or"]
temp_df.head()
```

```python
all_data = all_data[all_data["Order Date"].str[0:2] != "Or"]
all_data.head()
```

```python
all_data["Month"] = all_data["Order Date"].str[0:2]
all_data["Month"] = all_data["Month"].astype('int32')
all_data.head()
```

## Task 3: Add Sales Column

```python
all_data["Sales"] = all_data["Quantity Ordered"] * all_data["Price Each"]
all_data.head()
```

#### Convert columns to the correct type

```python
# Convert str fields to float and integer first way
all_data["Quantity Ordered"] = all_data["Quantity Ordered"].astype("int32") # make integer
all_data["Price Each"] = all_data["Price Each"].astype("float") # make float

# Second way
all_data["Quantity Ordered"] = pd.to_numeric(all_data["Quantity Ordered"])
all_data["Price Each"] = pd.to_numeric(all_data["Price Each"])
```

```python
all_data["Sales"] = all_data["Quantity Ordered"] * all_data["Price Each"]
all_data.head()
```

## Question 1: what was the best month for sales? How much was earned that month?

```python
results = all_data.groupby("Month").sum()
results.head(13)
```

```python
import matplotlib.pyplot as plt

months = range(1,13)

plt.bar(months, results["Sales"])
plt.xlabel("Month Number")
plt.ylabel("Sales in USD ($)")
plt.xticks(months)
plt.show()
```

## Question 2: What city had the highest number of sales

## Task 4: Add a City Column

```python
# create city column
# use apply function

# first way
all_data["City"] = all_data["Purchase Address"].apply(lambda x: x.split(",")[1])
all_data.head()

# second way
def get_city(address):
    return address.split(",")[1]
all_data["City"] = all_data["Purchase Address"].apply(lambda x: get_city(x))
all_data.head()
```

```python
# get the state

def get_state(address):
    return address.split(",")[2].split()[0]

all_data["City"] = all_data["Purchase Address"].apply(lambda x: get_city(x) + " " + get_state(x))
all_data.head()
```

```python
results = all_data.groupby("City").sum()
results
```

```python
cities = [city for city, df in all_data.groupby("City")]

plt.bar(cities, results["Sales"])
plt.xticks(cities, rotation="vertical")
plt.ylabel("Sales in USD($)")
plt.xlabel("Cities")
plt.show()
```

## Question 4: What products are most often sold together?

```python
# order id aynı olan ürünler beraber alınmış demektir
all_data.head()
```

```python
# burada keep first,last yada false secenekleri var. her iki duplicate i tutmak istiyorsak false demeliyiz
df = all_data[all_data["Order ID"].duplicated(keep=False)]
df.head()
```

```python
df = df.sort_values(["Order ID", "Product"])
df.head()
```

```python

df["Grouped"] = df.groupby("Order ID")["Product"].transform(lambda x: ','.join(x))

df = df[['Order ID','Grouped']].drop_duplicates()

df.head()
```

### VERY IMPORTANT COUNTING METHODS

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list = row.split(",")
    count.update(Counter(combinations(row_list,2)))
    
count.most_common(10)
```

## Question 5: What product sold the most? Why do you think it sold the most?

```python
product_group = all_data.groupby("Product")

quantity_ordered = product_group.sum()["Quantity Ordered"]
quantity_ordered

products = [product for product, df in product_group]
products

plt.bar(products, quantity_ordered)
plt.xticks(products, rotation="vertical")
plt.xlabel("Product Name")
plt.ylabel("Quantity Ordered")
plt.show()
```

```python
# her urunun fiyatını bulalım
# cunku ucuz oldugu icin cok satılıyor olabilir

prices = product_group.mean()["Price Each"]
prices

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered, color='y')
ax2.plot(products,prices,'b-')

ax1.set_xlabel("Product Name")
ax1.set_ylabel("Quantity Ordered", color='y')
ax2.set_ylabel("Price", color='b')
ax1.set_xticklabels(products,rotation='vertical')

plt.show()
```
