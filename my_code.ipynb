## 5. Manipulating Data

```python
# create a new column by using other columns
df["Total"] = df["HP"] + df["Attack"] + df["Defense"] + df["Sp. Atk"] + df["Sp. Def"] + df["Speed"]
df.sort_values("Total", ascending=False).head(5)
```

```python
# drop a column
df = df.drop(columns=["Total"])
df.head()
```

```python
# sum values
# bu sekilde yapmak her zaman güvenli olmayabilir
df["Total"] = df.iloc[:,4:10].sum(axis=1)
df.head()
```

```python
# change the position of the columns

cols = list(df.columns)
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]
df.head(5)
```

## 7. Filtering Data

```python
# çimen pokemonlarını getir
df.loc[df["Type 1"] == "Grass"]
```

```python
# çimen ve zehir pokemonlarını getir
df.loc[
    (df["Type 1"] == "Grass") 
    & 
    (df["Type 2"] == "Poison")
].sort_values("Total")
```

```python
# çimen ve zehir pokemonları ve HP 70 den buyuk pokemonları getir
new_df = df.loc[
    (df["Type 1"] == "Grass") 
    & 
    (df["Type 2"] == "Poison")
    &
    (df["HP"] > 70)
]
new_df
```

```python
# yeni index ekliyor ama eskisinide yeni bir column olarak tutuyor
new_df1 = new_df.reset_index()
new_df1
```

```python
# yeni index ekliyor,eskisini sildiriyo
new_df2 = new_df.reset_index(drop=True)
new_df2
```

```python
# yeni index ekliyor, eskisini siliyor ve aynı df icine yapıyor
new_df.reset_index(drop=True, inplace=True)
new_df
```

### Another Good Filtering Example about contains


```python
# isminde mega olan pokemonları getir
df.loc[df["Name"].str.contains("Mega")].head()
```

```python
# isminde mega olmayab pokemonları getir
df.loc[~df["Name"].str.contains("Mega")]
```

### Another filter example about contains and regex

```python
import re
df.loc[df["Type 1"].str.contains("Fire|Grass", regex=True)]
```

```python
# capital letter ignore
df.loc[df["Type 1"].str.contains("fire|grass", flags=re.I,regex=True)]
```

```python
# içinde "pi" geçen tüm pokemonları bul
df.loc[df["Name"].str.contains("pi[a-z]*",flags=re.I, regex=True)]
```

```python
# pi ile başlayan tüm pokemonları bul
df.loc[df["Name"].str.contains("^pi[a-z]*",flags=re.I, regex=True)]
```

## 9. Aggregate Statistics (Groupby)

```python
df.groupby(["Type 1"]).mean()
df.groupby(["Type 1"]).mean().sort_values("Defense", ascending=False).head(7)
```

```python
df.groupby(["Type 1"]).sum().head()
```

```python
df.groupby(["Type 1"]).count().head()
```

```python
df["count"] = 1
df.groupby(["Type 1"]).count()["count"].head()
```

```python
# iki grup sayma işlemi 
df.groupby(["Type 1","Type 2"]).count()["count"]
```

## 10. Working with Large Amount of Data

```python
new_df = pd.DataFrame(columns=[])
for df in pd.read_csv("modified.csv", chunksize=5):
    results = df.groupby(["Type 1"]).count()
    new_df = pd.concat([new_df,results])
new_df
```
